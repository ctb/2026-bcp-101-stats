# Basic statistical calculations

For a vector $x$ with $N$ elements,

## Average or mean

[Wikipedia: mean](https://en.wikipedia.org/wiki/Mean)

$$ \mu = \frac{1}{N} \sum_{i=1}^{N} x_i $$

## Standard deviation

[Wikipedia: standard deviation](https://en.wikipedia.org/wiki/Standard_deviation)

$$ \sigma = \sqrt{\sum_{i=1}^{N} (x_i - \mu)^2} $$

## Median

[Wikipedia: median ](https://en.wikipedia.org/wiki/Median)

if $N$ is odd:

$$ {\rm med(x)} = x_{(N+1)/ 2} $$

if $N$ is even, take the arithmetic mean of the two midpoint numbers:

$$ {\rm med(x)} = \frac{x_{N/2} + x_{N/2 + 1}}{2} $$
